```python
def grade_count(depart):
    juck1={'C':0,'B':0,'C+':0,
    'F':0, 
    'B+':0, 
    'A':0,'W':0}
    for i in range(6,24 ,3):
        
        increment1=dict(data[data['2']==depart][str(i)].value_counts())
        try:
            juck1["C"]=juck1["C"]+increment1["C"]
        except:
            juck1["C"]=juck1["C"]+0 
        try:
            juck1["B"]=juck1["B"]+increment1["B"]
        except:
            juck1["B"]=juck1["B"]+0 
        try:
            juck1["C+"]=juck1["C+"]+increment1["C+"]
        except:
            juck1["C+"]=juck1["C+"]+0 
        try:
            juck1["F"]=juck1["F"]+increment1["F"]
        except:
            juck1["F"]=juck1["F"]+0
        try: 
            juck1["W"]=juck1["W"]+increment1["W"]
        except:
            juck1["W"]=juck1["W"]+0
        try:
            juck1["B+"]=juck1["B+"]+increment1["B+"]
        except:
            juck1["B+"]=juck1["B+"]+0 
        try:
            juck1["A"]=juck1["A"]+increment1["A"] 
        except:
            juck1["A"]=juck1["A"]+0
    return juck1
```

```python
Dict_result={}
for i in ['MBA', 'BMS', 'BTD', 'M.SC(IR)', 'BBA-H', 'MS(SE)', 'BS(SE)',
    'BS(CS)', 'BSDS', 'BFD', 'MS(CS)', 'MS(BA)', 'BS(IR)', 'ADFD',
    'BS(A&F)', 'BSTN', 'MPHIL(IDS)', 'BS(ECO)', 'MPHIL(IR)', 'MS(MGT)',
    'BS(SC)', 'PH.D (CS)', 'FDFD', 'MSTN', 'FDTD', 'M.SC(DS)']:
    print("{0} => {1}".format(i,grade_count(i)))
    Dict_result[i]=grade_count(i )
```

```python
high_score_grade={}
for grade in ['A','B','B+','C','C+','F','W']:
    totel=data.loc[grade].sum()
    a=dict(data.loc[grade])
    dscore={}
    for i in a:
        #print('{0}       = {3}'.format(i,a[i],totel,a[i]/totel*100))
        dscore[i]=a[i]/totel*100
    max_val=max(dscore.values())
    deparment=list(dscore.keys())[list(dscore.values()).index(max(dscore.values()))]
    print('{0}  {1} :{2}'.format(grade,deparment,max_val))
    high_score_grade[grade]={deparment:max_val}
```
