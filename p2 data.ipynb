```python
def grade_calculator(dep,grad):
    s=data[(data['2']==dep) &(data['6']==grad)|(data['9']==grad)|(data['12']==grad)|(data['15']==grad)|(data['18']==grad)|(data['21']==grad)].shape
    return s[0] 
```

```python
Dict={}
for c,j in enumerate(['B+', 'B', 'C+', 'F', 'C', 'A', 'W']):
    for i in ['MBA', 'BMS', 'BTD', 'M.SC(IR)', 'BBA-H', 'MS(SE)', 'BS(SE)',
       'BS(CS)', 'BSDS', 'BFD', 'MS(CS)', 'MS(BA)', 'BS(IR)', 'ADFD',
       'BS(A&F)', 'BSTN', 'MPHIL(IDS)', 'BS(ECO)', 'MPHIL(IR)', 'MS(MGT)',
       'BS(SC)', 'PH.D (CS)', 'FDFD', 'MSTN', 'FDTD', 'M.SC(DS)']:
     #print(grade_calculator(i,j) )
     print('Degree {0} \n number of {1} =    {2}\n\n '.format(i,j,grade_calculator(i,j) ))
```

```python
Dict={}
for count,i in enumerate(['MBA', 'BMS', 'BTD', 'M.SC(IR)', 'BBA-H', 'MS(SE)', 'BS(SE)',
    'BS(CS)', 'BSDS', 'BFD', 'MS(CS)', 'MS(BA)', 'BS(IR)', 'ADFD',
    'BS(A&F)', 'BSTN', 'MPHIL(IDS)', 'BS(ECO)', 'MPHIL(IR)', 'MS(MGT)',
    'BS(SC)', 'PH.D (CS)', 'FDFD', 'MSTN', 'FDTD', 'M.SC(DS)']):
    dict_grad_counter={}
    for j in ['B+', 'B', 'C+', 'F', 'C', 'A', 'W']:
        dict_grad_counter[j]=grade_calculator(i,j)
    Dict[i]=dict_grad_counter
```
